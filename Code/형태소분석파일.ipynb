```python
import pandas as pd

df = pd.read_csv('news_nouns.csv', index_col='Unnamed: 0')
```

```python
df_group = df.groupby('month').agg({'nouns':'sum'})
```

```python
df_group.reset_index(inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def show_top30(text):
    tfidf = TfidfVectorizer(max_features=500, stop_words='english')
    tdm = tfidf.fit_transform(text)
    word_count = pd.DataFrame({
        '단어': tfidf.get_feature_names(),
        '빈도': tdm.sum(axis=0).flat
    })
    return word_count.sort_values('빈도', ascending=False)[:30]
```

```python
show_top30([df_group['nouns'][0]])
```

```python
show_top30([df_group['nouns'][1]])
```

```python
show_top30([df_group['nouns'][2]])
```

```python
show_top30([df_group['nouns'][3]])
```

```python
show_top30([df_group['nouns'][4]])
```
